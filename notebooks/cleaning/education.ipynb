# education_attainment_2003_2019

```python
df = pd.read_csv(config.r_education_data / 'education_attainment_2003_2019.csv')
df
```

```python
df.Education.unique()
```

```python
# format education column by removing the ISCED chunk
def remove_ISCED(edu):
    if ' - ' in edu:
        formatted_edu = edu.split(' - ')
        edu_cat = formatted_edu[0]
        return(edu_cat)
    return(edu)

df['Education'] = df['Education'].apply(remove_ISCED)
df
```

```python
df[df['Education'] == 'Total']
```

```python
# remove all rows where education == 'Total'
df = df[df['Education'] != 'Total']
df
```

```python
df['Age group/residence'].unique()
```

```python
# rename age/region column
df = df.rename(columns={'Age group/residence': 'age_region'})
```

```python
df = pd.melt(df, id_vars=['Sex', 'age_region', 'Education'], var_name='Year', value_name='Percent')
df
```

```python
# format percent column
df['Percent'] = df.Percent.apply(lambda x: round(x / 100, ndigits=3))
df
```

```python
# to csv
data_path = config.c_education_data / 'education_attainment_2003_2019.csv'
df.to_csv(data_path, index=False)
```

# field_of_study_1997_2019

```python
df = pd.read_csv(config.r_education_data / 'field_of_study_1997_2019.csv')
df
```

```python
cols = ['ISCED level', 'Type of education']
for col in cols:
    print(df[col].unique())
```

```python
# drop redundant cols
df = df.drop(columns=cols)
df
```

```python
df['Narrow field of study'].unique()
```

```python
df['Degrees'].unique()
```

```python
# rename cols
df = df.rename(columns={'Degrees': 'Degree', 'Narrow field of study': 'field'})
df
```

```python
# create male_female_total dimension
df = pd.melt(df, id_vars=['Year', 'field', 'Degree'], var_name='male_female_total', value_name='n_students_enrolled')
```

```python
# to csv
data_path = config.c_education_data / 'field_of_study_1997_2019.csv'
df.to_csv(data_path, index=False)
```
