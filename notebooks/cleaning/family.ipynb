```python
# try using mac_iceland encoding to read in the icelandic characters
df = pd.read_csv(config.r_family_data / 'family_type_1998_2020.csv', encoding='ISO-8859-1')
df
```

```python
df['Nuclear families'].unique()
```

```python
df = pd.melt(df, id_vars=['Nuclear families', 'Municipality'], var_name = 'family_type', value_name='n_families')
df
```

```python
# make year column
df['Year'] = df['family_type'].apply(lambda x: x.split()[0])

# remove the year from the family type column
def remove_year(cat):
    year_removed = cat.split()[1:]
    formatted_cat = ' '.join(year_removed)
    return(formatted_cat)

df['family_type'] = df['family_type'].apply(remove_year)

df
```

```python
# to csv
data_path = config.c_family_data / 'family_type_1998_2020.csv'
df.to_csv(data_path, index=False)
```
