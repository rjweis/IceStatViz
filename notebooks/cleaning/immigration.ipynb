# dx_citizenship_1991_2018

```python
df = pd.read_csv(config.r_immigration_data / 'dx_citizenship_1991_2018.csv')
df
```

* need year column, male/female/total column

```python
df = pd.melt(df, id_vars='Former citizenship', var_name='Year', value_name='n_icelandic_citizenship')
df
```

# migration_1986_2010

```python
df = pd.read_csv(config.r_immigration_data / 'migration_1986_2010.csv')
df
```

```python
for col in df.columns:
    print(col)
```

```python
# year to str
df.Year = df.Year.astype(str)
```

```python
df = pd.melt(df, id_vars=['Country', 'Year'], var_name='immigration_category', value_name='count')
df
```

```python
# change 'total total' to 'total
def rem_second_total(cat):
    if 'Total Total' in cat:
        split_words = cat.split()
        split_words.pop(-1)
        new_cat = ' '.join(split_words)
        return(new_cat)
    return(cat)

df['immigration_category'] = df['immigration_category'].apply(rem_second_total)
df
```

```python
df.info()
```

```python
# to csv
data_path = config.c_immigration_data / 'migration_1986_2010.csv'
df.to_csv(data_path, index=False)
```

# res_permits_2009_2018


```python
df = pd.read_csv(config.r_immigration_data / 'res_permits_2009_2018.csv')
df
```

```python
# need year, sex columns
df = pd.melt(df, id_vars=['Citizenship', 'Type'], var_name='Year', value_name='n_res_permits')
df
```

```python
# create sex column
df['Sex'] = df.Year.apply(lambda x: x.split()[1])

# modify year column
df['Year'] = df.Year.apply(lambda x: x.split()[0])

df
```

```python
# reorder cols
df = df[['Year', 'Citizenship', 'Sex', 'Type', 'n_res_permits']]
df
```

```python
# format missing data
df['n_res_permits'] = df['n_res_permits'].apply(lambda x: float('NaN') if x == '.' else int(x))
df
```

```python
df.info()
```

```python
# to csv
data_path = config.c_immigration_data / 'res_permits_2009_2018.csv'
df.to_csv(data_path, index=False)
```
