```python
import mypathsetup
import config
import pandas as pd
```

```python
df = pd.read_csv(config.r_tourism_data / 'kef_passengers_citizenship_2002_2020.csv')
df
```

```python
df = pd.melt(df, id_vars=['Citizenship'], var_name='Year', value_name='n_passengers')
df
```

```python
# remove missing values
df['n_passengers'] = df['n_passengers'].apply(lambda x: float('NaN') if x == '..' else x)
df
```

```python
# create month column
def int_to_month_str(date):
    int_str = date[-2:]

    if int_str == '01':
        return('Jan')
    if int_str == '02':
        return('Feb')
    if int_str == '03':
        return('Mar')
    if int_str == '04':
        return('Apr')
    if int_str == '05':
        return('May')
    if int_str == '06':
        return('Jun')
    if int_str == '07':
        return('Jul')
    if int_str == '08':
        return('Aug')
    if int_str == '09':
        return('Sep')
    if int_str == '10':
        return('Oct')
    if int_str == '11':
        return('Nov')
    if int_str == '12':
        return('Dec')

df['Month'] = df.Year.apply(int_to_month_str)

df
```

```python
# format year column
df.Year = df.Year.apply(lambda x: x[:4])
df
```

```python
# reorder columns
df = df[['Year', 'Month', 'Citizenship', 'n_passengers']]
df
```

```python
# to csv
data_path = config.c_tourism_data / 'kef_passengers_citizenship_2002_2020.csv'
df.to_csv(data_path, index=False)
```
