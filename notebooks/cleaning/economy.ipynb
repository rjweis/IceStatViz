# work_hours_by_sector_region_1991_2019

```python
df = pd.read_csv(config.r_economy_data / 'work_hours_by_sector_region_1991_2019.csv')
df
```

```python
df['Sex og Region'].unique()
```

```python
# rename columns
df = df.rename(columns={
        'Sex og Region': 'sex_region', 
        'Economic activity (main and second job)': 'sector'})
```

```python
df['sector'].unique()
```

```python
# remove the Capital letters prefix in the sectors
def remove_caps(val):
    split_vals = val.split(' - ')
    desc = split_vals.pop(-1)
    return(desc)

df['sector'] = df['sector'].apply(remove_caps)
df['sector'].unique()
```

```python
conf_cols = [col for col in df.columns if 'Conf.Int' in col]
total_cols = [col for col in df.columns if 'total' in col]
df = pd.melt(df, id_vars=['sex_region', 'sector'] + conf_cols, var_name='year', value_name='total_weekly_hrs')
df
```

```python
df = pd.melt(df, id_vars=['sex_region', 'sector', 'year', 'total_weekly_hrs'], value_name='.95_conf_int')
df
```

```python
# drop unnecessary cols, rename
df['year'] = df['year'].apply(lambda x: x.split()[0]) # take the year, the rest is not needed
df = df.drop(columns='variable')
df
```

```python
# rename the 'Activities of households as employers; u0ndifferentiated goods- and services-producing activitie' sector
def rename_cat(sector):
    if sector == 'Activities of households as employers; u0ndifferentiated goods- and services-producing activitie':
        new_sector = 'Activities of households as employers; undifferentiated goods- and services-producing activities'
        return(new_sector)
    return(sector)
df['sector'] = df['sector'].apply(rename_cat)
df['sector'].unique()
```

```python
# to csv
data_path = config.c_economy_data / 'work_hours_by_sector_region_1991_2019.csv'
df.to_csv(data_path, index=False)
```

# work_hours_schedule_2008_2017

```python
df = pd.read_csv(config.r_economy_data / 'work_hours_schedule_2008_2017.csv')
df
```

```python
df['Work-life balance'].unique()
```

* need male/female, local/immigrant dimensions
* need percent, hours, conf_int measures

Need to 'melt' three times - once for each measure

### pct column

```python
pct_cols = [col for col in df.columns if 'Percentage' in col]
hrs_cols = [col for col in df.columns if 'hours per week' in col]
conf_cols = [col for col in df.columns if 'Confidence interval' in col]
```

```python
pct_cols
```

```python
df = pd.melt(df, id_vars=['Year', 'Work-life balance'] + hrs_cols + conf_cols, value_name='percentage')
df
```

```python
# local/immigrant column
df['local_immigrant'] = df['variable'].apply(lambda x: 'local' if 'Locals' in x else 'immigrant')

# all/male/female columns
def to_all_male_female(cat):
    if 'Alls' in cat:
        return('all')
    if 'Males' in cat:
        return('male')
    if 'Females' in cat:
        return('female')

    return(cat)

df['male_female_all'] = df['variable'].apply(to_all_male_female)

df = df.drop(columns='variable')

df
```

### hours column

```python
df = pd.melt(df, id_vars=['Year', 'Work-life balance', 'percentage', 'local_immigrant', 'male_female_all'] + conf_cols, value_name='hours_per_week')
df = df.drop(columns='variable')
df
```

### conf int column

```python
id_cols = ['Year', 'Work-life balance', 'percentage', 'local_immigrant', 'male_female_all', 'hours_per_week']
df = pd.melt(df, id_vars=id_cols, value_name='.95_conf_int')
df = df.drop(columns='variable')
df
```

```python
# change .. to NaN
def to_nan(val):
    if val == '..':
        return(float('NaN'))
    return(val)

cols_to_modify = ['percentage', 'hours_per_week', '.95_conf_int']
for col in cols_to_modify:
    df[col] = df[col].apply(to_nan)

df
```

```python
data_path = config.c_economy_data / 'work_hours_schedule_2008_2017.csv'
df.to_csv(data_path, index=False)
```
